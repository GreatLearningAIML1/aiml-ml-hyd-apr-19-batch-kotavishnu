### Bank Customer churn prediction

Given a Bank customer, can we build a classifier which can determine whether they will leave or not using Neural networks?

### Load tensorflow

```python
import tensorflow as tf

tf.set_random_seed(42)
```

```python
tf.__version__
```

### Collect Data

```python
import keras
```

```python
#Read data from file
import pandas as pd
bank_df = pd.read_csv('./bank.csv',encoding='utf-8')
```

```python
bank_df.head()
```

```python
bank_df.describe().T
```

```python
#RowNumber and CustomerId are unique columns which may not add value to the information
#Droping these columns
bank_df=bank_df.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
bank_df.shape
```

```python
#Target column is Exited, Either the customer leave the bank or will not leave
X=bank_df.drop(['Exited'],axis=1)
y=bank_df['Exited']
```

```python
X.shape
```

```python
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

le=LabelEncoder()
X['Gender'] = le.fit_transform(X.Gender.values)
```

```python
le=LabelEncoder()
X['Geography'] = le.fit_transform(X.Geography.values)
```

```python
X.shape
```

```python
oh=OneHotEncoder(categorical_features=[1])
X=oh.fit_transform(X).toarray()
```

```python
X=X[:,1:]
```

```python
X
```

```python
#Splitting the data into train and test
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.24, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
model =Sequential()
model.add(Dense(output_dim=6,init="uniform",activation='relu',input_dim=11))
model.add(Dropout(rate=0.1))

model.add(Dense(output_dim=6,init="uniform",activation='relu',input_dim=11))
model.add(Dropout(rate=0.1))

model.add(Dense(output_dim=1,init="uniform",activation='sigmoid'))
```

```python
#Compile the model - add Loss and Gradient Descent optimizer
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,batch_size=10,nb_epoch=100)
```

```python
#Predict Test Values
y_pred=model.predict(X_test)
y_pred=(y_pred>0.5)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```
